```python
import osmnx as ox
```

```python
# Define the place names
place_names = ["Madura, East Java, Indonesia", "Surabaya, East Java, Java, Indonesia", "Gresik, East Java, Java, Indonesia", "Lamongan, East Java, Java, Indonesia", "Bojonegoro, East Java, Indonesia", "Tuban, East Java, Java, Indonesia"]

# Download and create the merged graph for the specified places
graph = ox.graph_from_place(place_names, network_type='drive')
```

```python
# Plot the graph
ox.plot_graph(graph, node_size=1)
```

```python
# Save the graph as a GraphML file
ox.save_graphml(graph, filepath='DATASET.graphml')
```
